### Data Preprocessing

```python
# Dropping unnecessary columns

fifa.drop(columns=['Photo','Flag','Club Logo'],axis=1,inplace=True)
```

```python
fifa.shape
```

```python
# defining a function to clean the object data type

def funct(Value):
    temp= Value.replace('€', '')
    if 'M' in temp:
        temp = float(temp.replace('M', ''))*1000000
    elif 'K' in Value:
        temp = float(temp.replace('K', ''))*1000
    return float(temp)
```

```python
# applying the function to value,wage column

fifa['Value'] = fifa['Value'].apply(lambda x: funct(x))
fifa['Wage'] = fifa['Wage'].apply(lambda x: funct(x))
```

```python
# Converting object to int of only year of Joined Column

fifa['Joined']=fifa['Joined'].fillna(0)
fifa['Joined']=pd.to_datetime(fifa['Joined'])
fifa['Joined']=fifa['Joined'].dt.year
```

```python
# converting str to datetime type of Contract Valid Until Column

fifa['Contract Valid Until']=pd.to_datetime(fifa['Contract Valid Until'])
```

```python
# converting objects into float of Height column.


fifa['Height']=fifa['Height'].astype(str).apply(lambda x:x.replace("'","."))
fifa['Height']=fifa['Height'].astype(float)

```

```python
# converting objects into float of Weight column.

fifa['Weight']=fifa['Weight'].astype(str).apply(lambda x:x.replace("lbs",""))
fifa['Weight']=fifa['Weight'].astype(float)

```

```python
# converting object into float of Release Clause by using def func

fifa['Release Clause'] = fifa['Value'].astype(str).apply(lambda x: funct(x))
```

```python
fifa.info()
```

```python
# Handling NaN 

fifa.isna().sum().sort_values(ascending=False)
```

```python
# filling NaN with mean values in the column of Height,Weight,Release Clause


fifa['International Reputation']=fifa['International Reputation'].fillna(fifa['International Reputation'].mean())
fifa['Height']=fifa['Height'].fillna(fifa['Height'].mean())
fifa['Weight']=fifa['Weight'].fillna(fifa['Weight'].mean())
fifa['Release Clause']=fifa['Release Clause'].fillna(fifa['Release Clause'].mean())
```
